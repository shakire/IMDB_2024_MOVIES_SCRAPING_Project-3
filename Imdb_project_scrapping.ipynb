```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time

driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31")

time.sleep(5)

genre_checkbox = driver.find_element(By.XPATH, '//*[@id="genreAccordion"]/div[1]/label/span[1]/div')
ActionChains(driver).move_to_element(genre_checkbox).click().perform()

time.sleep(5)

genre_selection = driver.find_elements(By.XPATH,'//*[@id="accordion-item-genreAccordion"]/div/section/button')
if genre_selection:
    ActionChains(driver).move_to_element(genre_selection[0]).click().perform()

time.sleep(5)

load_more = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button')
ActionChains(driver).move_to_element(load_more).click().perform()

time.sleep(5)

load_more = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button')
ActionChains(driver).move_to_element(load_more).click().perform()

time.sleep(3)

raw_title = driver.find_elements(By.CSS_SELECTOR,'h3[class="ipc-title__text"]')

time.sleep(3)

rating = driver.find_elements(By.CSS_SELECTOR,'span[class="ipc-rating-star--rating"]')

time.sleep(3)

voting_count = driver.find_elements(By.CSS_SELECTOR,'span[class="ipc-rating-star--voteCount"]') 

time.sleep(3)

durations = driver.find_elements(By.CSS_SELECTOR, 'div[class="sc-5179a348-6 bnnHxo dli-title-metadata"]')
```

```python
Title = []
for i in raw_title:
    text = i.text.strip()
    if '. ' in text:
        text = text.split('. ', 1)[1]
    Title.append(text)
Title.pop()
print(Title)

Ratings = []
for i in rating:
    Ratings.append(i.text)
print(Ratings)

Voting_count=[]
for i in voting_count:
    Voting_count.append(i.text)
print(Voting_count)

Duration=[]
for i in durations:
    Duration.append(i.text)
print(Duration)

import re 
timings = [re.search(r'\d+h(?: \d+m)?|\d+m', item).group() for item in Duration]
print(timings)
```

```python
print(len(Title))
print(len(Ratings))
print(len(Voting_count))
print(len(timings))
```

```python
import pandas as pd
df_action = pd.DataFrame({"Titile":Title,"Ratings":Ratings,"Voting_count":Voting_count,"Timings":timings})
df_action["Genre"] = "Action"
```

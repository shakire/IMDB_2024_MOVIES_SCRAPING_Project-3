```python
df["Voting_count"] = df["Voting_count"].str.replace(r"[()]", "", regex=True)
df["Voting_count"] = df["Voting_count"].str.replace("K", "", regex=False).astype(float) * 1000
```

```python
import re
def duration_to_minutes(duration):
    if pd.isna(duration):
        return None
    match = re.match(r'(?:(\d+)h)?\s*(?:(\d+)m)?', duration)
    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0
        return hours * 60 + minutes
    return None
```

```python
df["Timings"] = df["Timings"].apply(duration_to_minutes)
```
